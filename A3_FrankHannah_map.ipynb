```python
### Load libraries -----
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
import imageio
from matplotlib import patches
from matplotlib.lines import Line2D

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.linear_model import LinearRegression
```

```python
### Set font -----
# Source: https://matplotlib.org/stable/tutorials/text/usetex.html
plt.rcParams.update({
"text.usetex": True,
"font.family": "serif",
"font.sans-serif": "mathpazo",
})
```

```python
### Load Prio data ------
# Codebook: https://grid.prio.org/#/codebook
prio = pd.read_csv("prio.csv")
prio.head(3)
```

```python
### Multiple imputation of missing values --------
# https://www.numpyninja.com/post/mice-and-knn-missing-value-imputations-through-python
#lr = LinearRegression()
#imp = IterativeImputer(estimator=lr, max_iter=10, random_state=0)
#prio_imp = imp.fit_transform(prio)

#prio = pd.DataFrame(prio_imp, columns=prio.columns)
```

```python
### Check -----
# https://www.geeksforgeeks.org/working-with-missing-data-in-pandas/
prio.isnull().sum().any()
```

```python
# https://stackoverflow.com/questions/51070985/find-out-the-percentage-of-missing-values-in-each-column-in-the-given-dataset
missing = prio.isnull().sum() * 100 / len(prio)
missing_df = pd.DataFrame({'column_name': prio.columns,
                                 'percent_missing': missing})
missing_df.sort_values('percent_missing', inplace=True)
missing_df.head(30)
```

```python
### Alternatively just replace with mean ------
### https://machinelearningmastery.com/handle-missing-data-python/
prio["nlights_mean"].fillna(prio["nlights_mean"].mean(), inplace=True)
prio["prec_gpcc"].fillna(prio["prec_gpcc"].mean(), inplace=True)
```

```python
### Check -----
# https://www.geeksforgeeks.org/working-with-missing-data-in-pandas/
prio.isnull().sum().any()
```

```python
### Load Prio grid ------
prio_shp = gpd.read_file('priogrid_cell.shp')
prio_shp.head(3)
```

```python
### Merge Prio data with Prio grid ------
# Source: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html
df_prio = prio.merge(prio_shp, 
                     left_on='gid', 
                     right_on='gid', 
                     how="left")
df_prio.head(3)
```

```python
### Convert into geopandas df --------
# Source: https://geopandas.org/en/stable/gallery/create_geopandas_from_pandas.html
df_prio_geo = gpd.GeoDataFrame(
    df_prio, geometry=gpd.points_from_xy(df_prio.xcoord, df_prio.ycoord))
df_prio_geo.head(3)
```

```python
### Subset countries in Africa ---------
df_prio_geo_s = df_prio_geo[df_prio_geo["gwno"].isin([615, 540, 434, 571, 439, 516, 471, 482, 483, 581, 484, 522, 490,
       531, 530, 420, 452, 438, 404, 437, 501, 572, 570, 450, 620, 580,
       432, 435, 600, 541, 565, 436, 475, 517, 433, 451, 520, 560, 626,
       625, 510, 461, 616, 500, 551, 552, 651, 553])]
df_prio_geo_s
```

```python
### Load ucdp data -------
ucdp = pd.read_csv("https://ucdp.uu.se/downloads/ged/ged221-csv.zip", 
                  low_memory=False)
ucdp.head(3)
```

```python
### Get initial list of countries in Africa ------
list(ucdp["country_id"].loc[ucdp["region"]=="Africa"].unique())
```

```python
### Convert into geopandas df -------
# Source: https://geopandas.org/en/stable/gallery/create_geopandas_from_pandas.html
df_ucdp_geo = gpd.GeoDataFrame(
    ucdp, geometry=gpd.points_from_xy(ucdp.longitude, ucdp.latitude))
df_ucdp_geo.head(3)
```

```python
### Subset countries in Africa ---------
df_ucdp_geo_s = df_ucdp_geo[df_ucdp_geo["country_id"].isin([615, 540, 434, 571, 439, 516, 471, 482, 483, 581, 484, 522, 490,
       531, 530, 420, 452, 438, 404, 437, 501, 572, 570, 450, 620, 580,
       432, 435, 600, 541, 565, 436, 475, 517, 433, 451, 520, 560, 626,
       625, 510, 461, 616, 500, 551, 552, 651, 553])]
df_ucdp_geo_s.head(3)
```

```python
### (1) Distance to capital -------
for i in [1989.0,
 1990.0,
 1991.0,
 1992.0, 
 1993.0, 
 1994.0,
 1995.0,
 1996.0,  
 1997.0, 
 1998.0,
 1999.0, 
 2000.0, 
 2001.0,
 2002.0,
 2003.0,  
 2004.0, 
 2005.0, 
 2006.0,
 2007.0, 
 2008.0, 
 2009.0,
 2010.0, 
 2011.0, 
 2012.0,
 2013.0,
 2014.0]:
    
    fig, ax = plt.subplots(figsize = (10, 10))
    
    # https://stackoverflow.com/questions/9295026/how-to-remove-axis-legends-and-white-padding
    plt.axis('off')
    df_prio_geo_s.loc[(df_prio_geo_s["year"]==i)].plot(column='capdist',
                                                       ax=ax, 
                                                       marker='s', 
                                                       markersize=5, 
                                                       cmap="bone")
    
    df_ucdp_geo_s.loc[(df_ucdp_geo_s["year"]==i)].plot(column='best',
                                                       ax=ax, 
                                                       marker='o', 
                                                       markersize=df_ucdp_geo_s["best"]*1.5, 
                                                       color="mediumorchid", 
                                                       alpha=0.2)
    
    plt.title(f"Distance to capital and fatalities in battle for {int(i)}", size=15)
    
    #https://stackoverflow.com/questions/25408393/getting-individual-colors-from-a-color-map-in-matplotlib
    cmap = plt.cm.get_cmap('bone', 4)
    
    # https://matplotlib.org/stable/gallery/text_labels_and_annotations/custom_legends.html
    # https://stackoverflow.com/questions/40672088/matplotlib-customize-the-legend-to-show-squares-instead-of-rectangles
    entries1 = [patches.Rectangle((0,0),1,1, facecolor=cmap(0), label='0 km'),
                patches.Rectangle((0,0),1,1, facecolor=cmap(1), label='1000 km'),
                patches.Rectangle((0,0),1,1, facecolor=cmap(2), label='2000 km')]    
    
    # https://www.geeksforgeeks.org/use-multiple-columns-in-a-matplotlib-legend/
    # https://stackoverflow.com/questions/40672088/matplotlib-customize-the-legend-to-show-squares-instead-of-rectangles
    # https://stackoverflow.com/questions/25540259/remove-or-adapt-border-of-frame-of-legend-using-matplotlib
    # https://stackoverflow.com/questions/44620013/title-for-matplotlib-legend
    # https://stackoverflow.com/questions/72521182/how-to-reduce-space-between-columns-in-a-horizontal-legend-python
    # https://stackoverflow.com/questions/7125009/how-to-change-legend-size-with-matplotlib-pyplot
    legend1 = plt.legend(handles=entries1, 
               loc='lower center',
               bbox_to_anchor=(0.3,0.05),
               ncol=3,
               handlelength=1, 
               handleheight=1,
               columnspacing=0.8,
               frameon=False,
               title="Distance to capital (km)",
               prop={'size': 12})
    
    # Multiple legends: 
    # https://www.tutorialspoint.com/drawing-multiple-legends-on-the-same-axes-in-matplotlib
    plt.gca().add_artist(legend1)
    
    entries2 = [Line2D([0], [0], marker='o', color='w', markerfacecolor='mediumorchid', label='5', markersize=5),
               Line2D([0], [0], marker='o', color='w', markerfacecolor='mediumorchid', label='5000', markersize=10),
               Line2D([0], [0], marker='o', color='w', markerfacecolor='mediumorchid', label='30000', markersize=15)]
    
    plt.legend(handles=entries2, 
               loc='lower center', 
               ncol=3,
               bbox_to_anchor=(0.8,0.05),
               handlelength=1, 
               handleheight=1,
               columnspacing=0.8,
               frameon=False,
               title="Fatalities in battle",
               prop={'size': 12})

    plt.savefig(f"map_capdist_{int(i)}.jpeg", 
                dpi=400)
    
    plt.show()
```

```python
### Save as gif -------
# Source: https://towardsdatascience.com/probably-the-easiest-way-to-animate-your-python-plots-f5194ebed75f
with imageio.get_writer('animation_capdis.gif', mode='i') as writer:
    for i in [1989, 1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 
              1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006,
              2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014]:
        for n in range(6):
            image = imageio.imread(f'map_capdist_{i}.jpeg')
            writer.append_data(image)
```
